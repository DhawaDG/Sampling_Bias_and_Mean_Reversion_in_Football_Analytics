### Population Mean Vs, Sample mean

```python
import pandas as pd
dataset=pd.read_csv('premier-player-23-24.csv')
dataset.head()
```

```python
#lets try to find populations mean of expected goals and expected assist 
import numpy as np
population=dataset['xG']+dataset['xAG']

population_mean=population.mean()

print(f'Population mean of Combined Expected Goals and Expected Assist is: {population_mean:.2f}')
```

```python
#lets generate 5 sample mean and try to find there mean

np.random.seed(42)
sample_means=[population.sample(n=20).mean() for _ in range(5)]
mean=np.mean(sample_means)
print(f'sample mean list of 5 random(size=20) sample is:{sample_means}')
print(f'mean of sample means :{mean:.2f}')
```

```python
import matplotlib.pyplot as plt

# Assume these are already defined:
# population_mean
# sample_means

plt.figure(figsize=(10, 5))
plt.axhline(y=population_mean, color='red', linestyle='--', label=f'Population Mean :{population_mean:.2f}')  #,label=f'Random Ast Variance:{ast_var_season1:.2f}')
plt.axhline(y=mean,color='green',linestyle='--',label=f'sample mean :{mean:.2f}')

plt.scatter(range(1, 6), sample_means, color='blue', label='Sample Means (n=20 each)', s=100)

for i, mean in enumerate(sample_means, 1):
    plt.text(i, mean + 0.05, f"{mean:.2f}", ha='center', fontsize=9)

plt.title("Sample Means vs. Population Mean (xG + xAG)")
plt.xlabel("Sample Number")
plt.ylabel("Mean of xG + xAG")
plt.xticks(range(1, 6))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('C:/project/draft/Statistics_Project/Sampling_Bias_and_Mean_Reversion_in_Football_Analytics/visuals/Population_vs_sample_mean.png')
plt.show()
```
