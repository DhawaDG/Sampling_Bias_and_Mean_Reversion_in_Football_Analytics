### Unbiased vs. Biased Estimators


- Unbiased : Error is Random.
- Biased : Error is systematic.

```python
import pandas as pd
datasets=pd.read_csv('premier-player-23-24.csv')
datasets.head()
```

```python
#unbiased 
unbiased_sample=datasets['Gls_90']
unbiased_mean=unbiased_sample.mean()
print(f"Unbiased mean goals per 90: {unbiased_mean:.2f}")
```

```python
#biased 
biased_sample=datasets[datasets['Starts']>20]
biased_mean=biased_sample['Gls_90'].mean()
print(f"Biased mean goals per 90: {biased_mean:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot distributions
plt.figure(figsize=(10, 5))
sns.kdeplot(datasets['Gls_90'],label='All Players',fill=True)
sns.kdeplot(biased_sample['Gls_90'],label='Starters (>20 games)',fill=True)
plt.axvline(unbiased_mean,color='blue',linestyle='--',label='Unbiased Mean')
plt.axvline(biased_mean,color='red',linestyle='--',label='Biased Mean')
plt.xlabel('Goals per 90 Minutes')
plt.title('Biased vs Unbiased Goal Estimation')
plt.legend()
plt.savefig('C:/project/draft/Statistics_Project/Sampling_Bias_and_Mean_Reversion_in_Football_Analytics/visuals/Biased_Vs_Unbiased_sample.png')
plt.show()
```
